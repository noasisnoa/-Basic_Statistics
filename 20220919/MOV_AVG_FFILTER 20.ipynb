```python
import pandas
import numpy as np
import matplotlib.pyplot as plt
from scipy import io
```

```python
#데이터 불러오기
input_mat = io.loadmat('/content/drive/MyDrive/data/SonarAl.mat')
```

```python
# inqut_mat의 데이터를 보여준다.
print(input_mat)
```

```python
#sonarAlt 안에 있는 값을 알려준다.
print(input_mat['sonarAlt'])
```

```python
def get_sonar(i):
    """Measure sonar."""
    z = input_mat['sonarAlt'][0][i]  # input_mat['sonaralt']: (1, 1501)
    return z
```

```python
#인덱스는 0에서 부터 시작하기 때문에 n-1 을 해야 한다.
def mov_avg_filter(x_n, x_meas):
  n = len(x_n)
  for i in range(n-1):
    x_n[i] = x_n[i + 1]
    x_n[n-1] = x_meas
    x_avg = np.mean(x_n)

  return x_avg, x_n
```

```python
n_samples = 500

n = 10
time_end = 10

n1 = 20
time_end1 = 20
```

```python
dt = time_end / n_samples
time = np.arange(0, time_end, dt)
x_meas_save = np.zeros(n_samples)
x_avg_save = np.zeros(n_samples)

dt1 = time_end1 / n_samples
time1 = np.arange(0, time_end1, dt)
x_meas_save_20 = np.zeros(n_samples)
x_avg_save_20 = np.zeros(n_samples)
```

```python
print(time)
print(x_meas_save)
print(x_avg_save)

print(time1)
print(x_meas_save_20)
print(x_avg_save_20)
```

```python
for i in range(n_samples):
  x_meas = get_sonar(i)
  if i == 0:
    x_avg, x_n = x_meas, x_meas * np.ones(n) # x_meas * np.ones(n)뭐가 나오는지 확인하세요
  else:
   x_avg, x_n = mov_avg_filter(x_n, x_meas)

   x_meas_save[i] = x_meas
   x_avg_save[i] = x_avg   
```

```python
for i in range(n_samples):
  x_meas = get_sonar(i)
  if i == 0:
    x_avg, x_n = x_meas, x_meas * np.ones(n) # x_meas * np.ones(n)뭐가 나오는지 확인하세요
  else:
   x_avg, x_n = mov_avg_filter(x_n, x_meas)

   x_meas_save_20[i] = x_meas
   x_avg_save_20[i] = x_avg  
```

```python
plt.plot(time, x_meas_save, 'r*', label='Measured')
plt.plot(time, x_avg_save, 'b-', label='Moving average_10')
plt.plot(time, x_avg_save_20, 'g-', label='Moving average_20')

plt.legend(loc='upper left')
plt.title('Measured Altitudes v.s. Moving Average Filter Values')
plt.xlabel('Time [sec]')
plt.ylabel('Altitude [m]')
plt.savefig('/content/img')
```
