```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import io
import pandas as pd
```

```python
data = pd.read_csv('/content/drive/MyDrive/data/covid(1).csv', sep=',')
```

```python
print(data)
```

```python
def average(k, x_avg_k_1, x_measure):

  alpha = (k-1)/k
  x_avg_k = alpha * x_avg_k_1 + (1-alpha)*x_measure

  return x_avg_k
```

```python
n_samples = 960

time = np.arange(0, n_samples)
x_meas_save = np.zeros(n_samples)
x_avg_save = np.zeros(n_samples)
```

```python
x_avg_k_1 = 0

for i in range(0,960):
    k = i + 1
    x_measure = data
    x_avg_k = average(k, x_avg_k_1, x_measure) 

    x_meas_save = x_measure
    x_avg_save = x_avg_k    
    time[i] = k
    x_avg_k_1 = x_avg_k
```

```python
plt.plot(time, x_meas_save, 'r*', label='Measured')
plt.plot(time, x_avg_save, 'b-', label='avg')
plt.legend(loc='upper left')
plt.title('Measured Altitudes v.s. Moving Average Filter Values')
plt.xlabel('Time [sec]')
plt.ylabel('Altitude [m]')
plt.savefig('/content/img')
```
